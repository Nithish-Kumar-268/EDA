```python
def highBudget(group):
    return group['budget'].max()>= 100000000
```

```python
# risky? yes or no?
# director avg budget -> y
# movie budget > y
```

```python
def risky(group):
  group['risky'] = group['budget'] - group['budget'].mean() > 0
  return group
```

```python
data = data.groupby('director_name').apply(risky)
```

```python
# active years for a director
```

```python
data_agg = data.groupby('director_name').aggregate({'year': ['min', 'max'], 'title': 'count'})
data_agg
```

```python
data_agg.head()
```

```python
data_agg.columns
```

```python
data_agg.columns = ['_'.join(col) for col in data_agg.columns]
```

```python
data_agg.head()
```

```python
data_agg = data_agg.reset_index()
```

```python
data_agg.head()
```

```python
data_agg["Active_years"] = data_agg['year_max'] - data_agg['year_min']
```

```python
data_agg.head()
```

```python
data_agg['movies_per_year'] = (data_agg['title_count']/data_agg['Active_years']).round(2)
data_agg.head()
```

```python
data_agg = data_agg.sort_values(by='movies_per_year' , ascending =False)
data_agg
```
